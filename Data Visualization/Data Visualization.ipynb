```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Line Plot

```python
line_plot = pd.read_csv('spotify.csv')
```

```python
line_plot.head()
```

```python
line_plot.tail()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=line_plot)
plt.title('Spotify songs line graph')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=line_plot['Shape of You'])
sns.lineplot(data=line_plot['Unforgettable'])
plt.title('Spotify songs line graph')
plt.show()
```

#### Bar Plot

```python
bar_plot = pd.read_csv('flight_delays.csv')
```

```python
bar_plot.head()
```

```python
bar_plot.tail()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x = bar_plot.index,y = bar_plot['AA'])
plt.title('Flight history for American Airline')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x = bar_plot.index,y = bar_plot['AS'])
plt.title('Flight history for Alaska Airline')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(data = bar_plot,annot=True)
plt.title('Heat Map')
plt.show()
```

#### Scatter Plot

```python
scatter_plot = pd.read_csv('insurance.csv')
```

```python
scatter_plot.head()
```

```python
scatter_plot.tail()
```

```python
sns.scatterplot(x=scatter_plot['bmi'],y=scatter_plot['charges'])
```

```python
sns.regplot(x=scatter_plot['bmi'],y=scatter_plot['charges'])
```

```python
sns.scatterplot(x=scatter_plot['bmi'],y=scatter_plot['charges'],hue=scatter_plot['smoker'])
```

```python
sns.lmplot(x='bmi',y='charges',hue='smoker',data = scatter_plot)
```

```python
sns.swarmplot(x = scatter_plot['smoker'],y=scatter_plot['charges'])
```

#### Histogram

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris.tail()
```

```python
sns.distplot(a = iris['Sepal Length (cm)'],kde=False)
sns.distplot(a = iris['Sepal Width (cm)'],kde=False)
```

```python
sns.kdeplot(data = iris['Petal Length (cm)'],shade=True)
sns.kdeplot(data = iris['Petal Width (cm)'],shade=True)
```

```python
sns.jointplot(x=iris['Petal Length (cm)'],y=iris['Sepal Width (cm)'],kind='kde')
```
